```python
#problem 1
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

def sa(x):
    return (x - 1) * np.log(x) - x + 0.5 * np.log(2 * np.pi * x)

x = np.linspace(2, 100)
y2 = np.log(x)

x = np.linspace(2, 100)
y1 = sa(x)

plt.figure(figsize=(8, 6))  
plt.plot(x, y2, label='ln(x)') 
plt.plot(x, y1, label = 'stirlings approximation')
plt.xlabel('n')  
plt.legend() 
plt.show()  
```
