```python
#problem 1
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

def sa(x):
    return (x - 1) * np.log(x) - x + 0.5 * np.log(2 * np.pi * x)

x = np.linspace(2, 100)
y2 = np.log(x)

x = np.linspace(2, 100)
y1 = sa(x)

plt.figure(figsize=(8, 6))  
plt.plot(x, y2, label='ln(x)') 
plt.plot(x, y1, label = 'stirlings approximation')
plt.xlabel('n')  
plt.legend() 
plt.show()  
```

```python
#problem 5 part a and b
import numpy as np
import matplotlib.pyplot as plt

rows, cols = 1000, 1000 #makes matrix
matrix = np.random.randn(rows, cols) 
print(matrix.shape)

fig, axes = plt.subplots(1, 3, figsize=(12, 6)) #to makw the subplots

axes[0].hist(matrix[0, :], bins =30) #histogram 1
axes[0].set_xlabel('x random values')

axes[1].hist(matrix[1, :], bins=30) #histogram 2
axes[1].set_xlabel('y random values')

x = matrix[0, :]
y = matrix[1, :]
hist2d, xedge, yedge = (np.histogram2d)(x, y, bins = 30)

lol = axes[2].imshow(hist2d.T)
axes[2].set_xlabel('heatmap')
plt.colorbar(lol, ax = axes[2])

plt.show()
```
